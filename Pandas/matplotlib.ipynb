```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.ylabel("some numbers y")
plt.xlabel("some numbers x")
plt.show()
```

```python
x,y = [2,6,10,14], [1,5,9,13]
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
