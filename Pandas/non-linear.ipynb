```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit            
```

```python
np.random.seed(0)
x = np.linspace(-5,5, num=50)
y = 2+1.5*x+3*x**2+np.random.normal(scale=2.0, size=x.shape)
```

```python
np.random.normal(scale=10.0)
```

```python
plt.figure(figsize=(4,3))
plt.scatter(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.show()
```

```python
# Define the nonlinear regression func

def quadratic_func(x,a,b,c):
    return a*x**2+b*x+c
```

```python
popt, pcov = curve_fit(quadratic_func, x,y)
```

```python
#Visualize the results

plt.scatter(x,y, label='Data')
plt.plot(x, quadratic_func(x, *popt), 'r-', label="Fit")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pds
import matplotlib.pyplot as plot
df = pds.read_csv('china_gdp.csv')
df.head(10)
```

```python
plot.figure(figsize=(8,3))
x_data = df['Year'].values
y_data = df['Value'].values
plot.plot(x_data, y_data, 'r^')
plot.ylabel('GDP')
plot.xlabel('Year')
plot.show()
```

```python
#sigmoid func   
x = np.arange(-5.0,5.0, 0.1)
y = 1.0/(1.0+np.exp(-x))
plot.plot(x,y)
plot.ylabel('Dependent Variable')
plot.xlabel('Independent Variable')
plot.show()
```

```python
def sigmoid(x, Beta_1, Beta_2):
    y = 1/(1+np.exp(-Beta_1*(x-Beta_2)))
    return y
```

```python
xdata = x_data/max(x_data)
ydata = y_data/max(y_data)
popt, pcov = curve_fit(sigmoid, xdata, ydata)

x = np.linspace(1960, 2015, 55)
x = x/max(x)

plot.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plot.plot(xdata, ydata, 'ro', label='data')
plot.plot(x,y, linewidth=3.0, label='fit')
plot.legend(loc='best')
plot.ylabel('GDP')
plot.xlabel('Year')
plot.show()
```
