```python
import pandas as pd 
import matplotlib 
import scipy
import seaborn as sns 
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset("car_crashes")
df.head(10)
```

```python
from matplotlib import pyplot as plt
x = df.speeding
y = df.alcohol
plt.scatter(x,y)
plt.xlabel("Speeding")
plt.ylabel("Alcohol")
plt.show()
```

```python
plt.scatter("speeding", 'alcohol', data=df)
sns.set_style("darkgrid", {"grid.color": '0.5'})
sns.despine()
plt.show()
```

```python
df = sns.load_dataset("iris")
sns.histplot(df['petal_length'])
```

```python
tips = sns.load_dataset('tips')

tips.time.replace('Dinner', 1, inplace=True)
tips.time.replace('Lunch', 2, inplace=True)

plt.scatter(x=tips['total_bill'], y=tips['size'])
plt.xlabel("tip")
plt.ylabel("size")
plt.show()

```

```python
tips = sns.load_dataset('tips')

tips.day.replace('Sun', 3, inplace=True)
tips.day.replace('Sat', 2, inplace=True)
tips.day.replace('Thur', 0, inplace=True)
tips.day.replace('Fri', 1, inplace=True)

plt.scatter(x=tips['total_bill'], y=tips['day'])
plt.xlabel("tip")
plt.ylabel("day")
plt.show()
```
